## MODELIZAR

### Reservar el dataset de validacion

```python
train_x,val_x,train_y,val_y = train_test_split(x,y,test_size=0.3)
```

### Crear el pipe y el diccionario de algorimos, parámetros y valores a testar

Usaremos LGBM en vez de una Regresión Linear Múltiple porque la predicción de RL no era útil

No realizamos preselección de variables porque son muy pocas con las que trabajamos y porque los modelos de árboles no es necesario para que predigan bien

```python
pipe = Pipeline([('algoritmo',Ridge())])

grid = [      
        {'algoritmo': [Ridge()],
         'algoritmo__alpha': list(np.arange(0.1,1.1,0.1))},
        
        {'algoritmo': [Lasso()],
         'algoritmo__alpha': list(np.arange(0.1,1.1,0.1))},
    
        {'algoritmo': [HistGradientBoostingRegressor(min_samples_leaf = 100, scoring = 'neg_mean_absolute_percentage_error')],
         'algoritmo__learning_rate': [0.01,0.025,0.05,0.1],
         'algoritmo__max_iter': [50,100,200],
         'algoritmo__max_depth': [5,10,20],
         'algoritmo__l2_regularization': [0,0.25,0.5,0.75,1]}
       ]
```

### Optimizar los hiper parámetros

####  Con grid search

```python
grid_search = GridSearchCV(estimator= pipe, 
                           param_grid = grid, 
                           cv = 3, 
                           scoring = 'neg_mean_absolute_error', #Xq mediremos el error en % de error
                           verbose = 0,
                           n_jobs = -1)

modelo = grid_search.fit(train_x,train_y)
```

```python
#Mostrar mas caracteres de las columnas. Se usa cuando se corta el texto
pd.set_option('display.max_colwidth', None) #Número de caractres que deben verse. None = Máx

salida = pd.DataFrame(grid_search.cv_results_).sort_values(by = 'rank_test_score').reset_index(drop=True).head(10)
salida.T #Yo lo saco traspuesto porque veo mejor los parámetros que tengo que mirar que son la primera columna

#mean_test_score nos dice el % de +- error que estamos cometiendo porque usa neg_mean_absolute_percentage_error.
#Ej: mean_test_score: 21 quiere decir que si el modelo predice que un cliente va a tener default en el 50% de amortización
#del préstamo. Podríamos decir que el default estará entre 50% +- 21%: (29%-71%)
```

Vemos que el resultado es el mismo en todas las primeras configuraciones.

Por sencillez nos quedamos con el primero.

Y reentreamos el modelo con esa configuración.

```python
modelo_ead = HistGradientBoostingRegressor(learning_rate = 0.1,
                                          max_iter = 200,
                                          max_depth = 10,
                                          min_samples_leaf = 100,
                                          scoring = 'neg_mean_absolute_percentage_error',
                                          l2_regularization = 0.75)
modelo_ead.fit(train_x,train_y)
```

## EVALUAR

### Predecir sobre validación

```python
pred = modelo_ead.predict(val_x)
```

Corregimos los máximos y mínimos xq son porcentajes

```python
pred = np.where(pred < 0, 0, pred)
pred = np.where(pred > 1, 1, pred)
```

### Evaluar sobre validación

```python
mean_absolute_error(val_y, pred)
```
